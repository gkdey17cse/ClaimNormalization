Plot 1 : Epoch vs Train & Validation Error

```python
import matplotlib.pyplot as plt

# --- Data ---
epochs = [1, 2, 3, 4, 5]

# BART-Base
bart_train_loss = [2.0445, 0.4888, 0.4124, 0.3522, 0.2990]
bart_val_loss = [0.4986, 0.4822, 0.4738, 0.4775, 0.4912]

# T5-Small
t5_train_loss = [3.2131, 2.9375, 2.8765, 2.8211, 2.8024]
t5_val_loss = [2.9105, 2.8050, 2.7602, 2.7399, 2.7329]

# --- Plotting ---
plt.figure(figsize=(10, 6))

# BART
plt.plot(epochs, bart_train_loss, label="BART Train Loss", marker='o', linestyle='-')
plt.plot(epochs, bart_val_loss, label="BART Val Loss", marker='o', linestyle='--')

# T5
plt.plot(epochs, t5_train_loss, label="T5 Train Loss", marker='s', linestyle='-')
plt.plot(epochs, t5_val_loss, label="T5 Val Loss", marker='s', linestyle='--')

# --- Formatting ---
plt.title("Epoch vs Train/Val Loss for BART-Base and T5-Small")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend()
plt.tight_layout()
plt.show()
```

Plot 2 : Epoch vs BLEU , BARTscore & ROGUE-L

```python
import matplotlib.pyplot as plt

# Epochs
epochs = [1, 2, 3, 4, 5]

# --- BART-Base Scores ---
bart_rouge = [0.3368, 0.3351, 0.3380, 0.3356, 0.3310]
bart_bleu = [0.1911, 0.1886, 0.1913, 0.2066, 0.2139]
bart_bert = [0.8807, 0.8802, 0.8809, 0.8803, 0.8819]

# --- T5-Small Scores ---
t5_rouge = [0.3127, 0.3154, 0.3179, 0.3210, 0.3195]
t5_bleu = [0.1422, 0.1581, 0.1659, 0.1704, 0.1687]
t5_bert = [0.8742, 0.8751, 0.8760, 0.8767, 0.8771]

# --- Plotting ---
fig, axs = plt.subplots(3, 1, figsize=(10, 10), sharex=True)

# ROUGE-L
axs[0].plot(epochs, bart_rouge, marker="o", label="BART ROUGE-L")
axs[0].plot(epochs, t5_rouge, marker="s", label="T5 ROUGE-L")
axs[0].set_ylabel("ROUGE-L")
axs[0].legend()
axs[0].grid(True, linestyle="--", alpha=0.6)

# BLEU-4
axs[1].plot(epochs, bart_bleu, marker="o", label="BART BLEU-4")
axs[1].plot(epochs, t5_bleu, marker="s", label="T5 BLEU-4")
axs[1].set_ylabel("BLEU-4")
axs[1].legend()
axs[1].grid(True, linestyle="--", alpha=0.6)

# BERTScore
axs[2].plot(epochs, bart_bert, marker="o", label="BART BERTScore")
axs[2].plot(epochs, t5_bert, marker="s", label="T5 BERTScore")
axs[2].set_ylabel("BERTScore")
axs[2].set_xlabel("Epoch")
axs[2].legend()
axs[2].grid(True, linestyle="--", alpha=0.6)

# --- Final Touch ---
plt.suptitle(
    "Epoch vs ROUGE-L / BLEU-4 / BERTScore for BART-Base and T5-Small", fontsize=14
)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```
